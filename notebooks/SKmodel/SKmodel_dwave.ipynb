### ã‚¬ã‚¦ã‚¹åˆ†å¸ƒã‹ã‚‰ç›¸äº’ä½œç”¨ã®ä½œæˆ

ã¾ãšã€ã‚¬ã‚¦ã‚¹åˆ†å¸ƒã‹ã‚‰ç›¸äº’ä½œç”¨$J$ï¼ˆã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ï¼‰ã‚’ï¼‘ã¤ç”Ÿæˆã—ã¾ã™ã€‚ã“ã“ã§ã€ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã¯ä¸Šè¨˜ã—ãŸé€šã‚Šã€

|  $\sigma$  | $2/\sqrt{ğ‘}$ | æ¨™æº–åå·® |


ã¨ã—ã¦ã„ã¾ã™ã€‚

```python
def J_generator(N, mu, sigma):
    '''ç¢ºç‡åˆ†å¸ƒã‹ã‚‰ç›¸äº’ä½œç”¨ã‚’ä½œæˆ'''
    J = {(i, j): np.random.normal(loc=mu, scale=sigma) for i in range(N) for j in range(i+1, N)}
    return J
```

```python
ans_J = J_generator(N=15, mu=0, sigma=2/np.sqrt(15))
```

### ç›¸äº’ä½œç”¨ã‹ã‚‰ã‚¹ãƒ”ãƒ³é…ä½ã®ä½œæˆ

æ¬¡ã«ã€ä½œæˆã—ãŸç›¸äº’ä½œç”¨$J$ã‚’ç”¨ã„ã¦ã€SKæ¨¡å‹ã®ã‚¹ãƒ”ãƒ³é…ä½ã‚’150å€‹ã‚µãƒ³ãƒ—ãƒªãƒ³ã‚°ã—ã¾ã™ã€‚

ã“ã“ã§ã¯ã€ã‚µãƒ³ãƒ—ãƒªãƒ³ã‚°ã«MCMC(ãƒ¡ãƒˆãƒ­ãƒãƒªã‚¹æ³•)ã‚’åˆ©ç”¨ã—ã¾ã™ã€‚æ‰‹é †ã¯æ¬¡ã®é€šã‚Šã§ã™ã€‚



> 1.   å…¨ã‚µã‚¤ãƒˆã‹ã‚‰ã‚ã‚‹ï¼‘ã‚µã‚¤ãƒˆã‚’ä¹±æ•°ã«ã‚ˆã‚Šé¸æŠã—ã€ ãã®ã‚µã‚¤ãƒˆã®ã‚¹ãƒ”ãƒ³ã®å‘ãã‚’åè»¢ã•ã›ãŸçŠ¶æ…‹ã‚’è©¦è¡ŒçŠ¶æ…‹ã¨ã™ã‚‹
> 2.   è©¦è¡ŒçŠ¶æ…‹ã«ãŠã‘ã‚‹ã‚¨ãƒãƒ«ã‚®ãƒ¼ã¨å…ƒã®çŠ¶æ…‹ã®ã‚¨ãƒãƒ«ã‚®ãƒ¼ã®å·®$\Delta E$ã‚’æ±‚ã‚ã‚‹ã€‚
> 3.   ä»¥ä¸‹ã®åˆ¤å®šã«ã‚ˆã‚Šã€è©¦è¡ŒçŠ¶æ…‹ã‚’æ–°ãŸãªçŠ¶æ…‹ã¨ã™ã‚‹ã‹ã€ã‚‚ã¨ã®çŠ¶æ…‹ã‚’ç¶­æŒã™ã‚‹ã‹ã‚’æ±ºã‚ã‚‹ã€‚
> *   $\Delta E < 0$ã®å ´åˆã«ã¯ã€è©¦è¡ŒçŠ¶æ…‹ã‚’æ–°ãŸãªçŠ¶æ…‹ã¨ã™ã‚‹ã€‚
> *   ä¸€æ–¹ã€$\Delta E > 0$ã®å ´åˆã«ã¯ã€ ä¹±æ•°$a$ ã‚’ $0â‰¤aâ‰¤1$ ã®ç¯„å›²ã§ç”Ÿæˆã™ã‚‹ã€‚ $a<\exp(-\beta \Delta E)$ ã®å ´åˆã«ã¯ã€ è©¦è¡ŒçŠ¶æ…‹ã‚’æ–°ãŸãªçŠ¶æ…‹ã¨ã™ã‚‹ã€‚ãã†ã§ãªã„å ´åˆã«ã¯ã‚‚ã¨ã®çŠ¶æ…‹ã‚’ç¶­æŒã™ã‚‹ã€‚
>
> 4.   1~3ã‚’ç¹°ã‚Šè¿”ã™ã€‚
>
> å¼•ç”¨å…ƒï¼šhttp://www.cmpt.phys.tohoku.ac.jp/~izumida/Ising/IsingMCMP.html

```python
def cal_energy(N, J, spin):
    energy = np.sum([-J[(i,j)] * spin[i] * spin[j] for i in range(N) for j in range(i+1, N)])
    return energy

def MCMC(N, J, iter, beta=1):
    # ãƒ©ãƒ³ãƒ€ãƒ ã«ã‚¹ãƒ”ãƒ³é…ä½ã‚’ä½œã‚‹
    minS = np.random.choice([-1, 1], size=N)
    # åˆæœŸçŠ¶æ…‹ã®ã‚¨ãƒãƒ«ã‚®ãƒ¼ã‚’æ±‚ã‚ã‚‹
    minE = cal_energy(N=N, J=J, spin=minS)

    for n in range(iter):
        # åè»¢ã™ã‚‹ã‚¹ãƒ”ãƒ³ã‚’ãƒ©ãƒ³ãƒ€ãƒ ã«æ±ºå®š
        i = np.random.randint(0, N)

        # ï¼‘ã¤ã®ã‚¹ãƒ”ãƒ³ã‚’åè»¢ã•ã›ãŸã‚¹ãƒ”ãƒ³é…ä½ã‚’ä½œã‚‹
        flipS = minS.copy()
        flipS[i] *= -1

        # åè»¢å¾Œã®ã‚¨ãƒãƒ«ã‚®ãƒ¼ã‚’æ±‚ã‚ã‚‹
        flipE = cal_energy(N=N, J=J, spin=flipS)

        # åè»¢å‰ã¨å¾Œã®ã‚¨ãƒãƒ«ã‚®ãƒ¼ã®å·®
        deltaE = flipE - minE

        if deltaE < 0: # ã‚¨ãƒãƒ«ã‚®ãƒ¼ã®å°ã•ã„æ–¹ã‚’æ¡ç”¨ã™ã‚‹
            minE = flipE
            minS = flipS.copy()
        else:
            a = random.random()
            p = np.exp(- beta * deltaE)
            if a < p:
                minE = flipE
                minS = flipS.copy()
            else:
                pass

    return minS
```

æº–å‚™ãŒçµ‚ã‚ã‚Šã¾ã—ãŸã€‚

150å€‹ã®ã‚¹ãƒ”ãƒ³é…ä½ã‚’ä½œæˆã—ã¾ã—ã‚‡ã†ã€‚

```python
def MCMC_sampling(num_instances):
    sample_lists = []
    for i in range(num_instances):
        sample = MCMC(N=15, J=ans_J, iter=100)
        sample_lists.append(sample)
    return sample_lists
```

```python
data_lists = MCMC_sampling(num_instances=150)
```

ã“ã‚Œã§ã€ç›¸äº’ä½œç”¨$J$ã‹ã‚‰ã‚¹ãƒ”ãƒ³é…ä½ã‚’ä½œæˆã§ãã¾ã—ãŸã€‚

### æ­£è§£ã¨ãªã‚‹å¹³å‡å¯¾æ•°å°¤åº¦ã®è¨ˆç®—

ã“ã“ã§ã¯ã€ä½œæˆã—ãŸã‚¹ãƒ”ãƒ³é…ä½ã«ã‚ˆã‚‹å¹³å‡å¯¾æ•°å°¤åº¦ã‚’è¨ˆç®—ã—ã¾ã™ã€‚å¹³å‡å¯¾æ•°å°¤åº¦ã¯æ¬¡ã§è¡¨ã›ã¾ã—ãŸã€‚

$$
\begin{aligned}
\Lambda_{\mathrm{av}}(S) &=\frac{1}{L} \sum_{l=1}^{L} \log Q\left(\mathbf{s}^{(l}\right) \\
&=-\beta \frac{1}{L} \sum_{l=1}^{L} E\left(\mathbf{s}^{(l)}\right)-\log Z(\beta)
\end{aligned}
$$

ã¾ãšã€å¹³å‡å¯¾æ•°å°¤åº¦ã®ç¬¬ï¼‘é …éƒ¨åˆ†ã‚’æ±‚ã‚ã¾ã™ã€‚

```python
def cal_first(beta, J, data_lists):
    energy_ave = np.average([cal_energy(N=len(data_list), J=J, spin=data_list) \
                             for data_list in data_lists])
    first = - beta * energy_ave
    return first
```

```python
data_first = cal_first(beta=1, J=ans_J, data_lists=data_lists)
print(data_first)
```

ç¬¬ï¼‘é …ãŒè¨ˆç®—ã§ãã¾ã—ãŸã€‚ç¶šã„ã¦ã€å¹³å‡å¯¾æ•°å°¤åº¦ã®ç¬¬ï¼’é …ã‚’æ±‚ã‚ã¾ã™ã€‚

```python
def cal_second(beta, N, J):
    state_sum = np.sum([ np.exp(- beta * cal_energy(N=N, J=J, spin=z)) \
                        for z in itertools.product([-1,1], repeat=N)])
    second = np.log(state_sum)
    return second
```

```python
data_second = cal_second(beta=1, N=15, J=ans_J)
print(data_second)
```

```python
# æ•™å¸«ãƒ‡ãƒ¼ã‚¿ã®å¹³å‡å¯¾æ•°å°¤åº¦
data_LH = data_first - data_second
print(data_LH)
```

### QAã«ã‚ˆã‚‹å­¦ç¿’

ã¾ãšåˆã‚ã«ã€QAã«ã‚ˆã‚‹å­¦ç¿’ã‚’è¡Œã„ã¾ã™ã€‚ä½¿ç”¨ã™ã‚‹å‹¾é…æ³•ã‚’ã‚‚ã†ä¸€åº¦è¨˜è¼‰ã—ã¾ã™ã€‚

$$
J_{i j}(t+1)=J_{i j}(t)+\eta \frac{\partial S}{\partial J_{i j}}
$$

$$
\frac{1}{\beta} \frac{\partial S}{\partial J_{i j}}=\left\langle s_{i} s_{j}\right\rangle_{\rho _\mathcal{D}}-\left\langle s_{i} s_{j}\right\rangle_{\rho}
$$

ã“ã“ã§ã€æœŸå¾…å€¤$\left\langle{}\right\rangle_{\rho}$ã¯

$$
\left\langle s_{i} s_{j}\right\rangle_{\rho} = \frac{1}{L} \sum_{l=1}^{L} s_{i}^{(l)} s_{j}^{(l)}
$$

ã®ã‚ˆã†ã«ã€D-Waveãƒã‚·ãƒ³ã‚’ã‚µãƒ³ãƒ—ãƒ©ãƒ¼ã¨ã—ã¦ç”¨ã„ã‚‹ã“ã¨ã§è¿‘ä¼¼çš„ã«æ±‚ã‚ã‚‰ã‚Œã¾ã™ã€‚

```python
def minus_J(num_spin, J):
    '''ã‚³ã‚¹ãƒˆé–¢æ•°ã®ç›¸äº’ä½œç”¨ã«ãƒã‚¤ãƒŠã‚¹ãŒã¤ã„ã¦ã„ã‚‹ãŸã‚ã€ãã®ã‚ˆã†ã«å®Ÿè£…ã—ã¾ã™'''
    m_J = {(i, j): J[(i, j)] * -1 for i in range(num_spin) for j in range(i+1, num_spin)}
    return m_J

def sampling(method, h, J, Nsample, beta):
    if method == 'QA':
        sampleset = sampler.sample_ising(h=h, J=J, num_reads=Nsample, beta=beta, answer_mode='raw', postprocess='sampling') 
    if method == 'SA':
        sampleset = sampler.sample_ising(h=h, J=J, num_reads=Nsample, beta_min=beta, beta_max=beta)
    return sampleset

def train_model(method, Tall, Nsample, beta, eta, data_lists, sampler):
    num_data=len(data_lists) # ãƒ‡ãƒ¼ã‚¿ã®æ•°
    num_spin=len(data_lists[0]) # ã‚¹ãƒ”ãƒ³ã®æ•°

    # ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã®åˆæœŸå€¤
    train_J= {(i, j): 0 for i in range(num_spin) for j in range(i+1, num_spin)}
    h = {i: 0 for i in range(num_spin)}

    # ãƒ‡ãƒ¼ã‚¿ã®çµŒé¨“å¹³å‡
    data_ave={}
    for i in range(num_spin):
        for j in range(i+1, num_spin):
            data_ave[(i, j)] = np.average([data_lists[k][i] * data_lists[k][j] for k in range(len(data_lists))])

    LH_hists={}
    for t in tqdm(np.arange(Tall)):
        # ã‚µãƒ³ãƒ—ãƒªãƒ³ã‚°
        m_J = minus_J(num_spin=num_spin, J=train_J)
        sampleset = sampling(method, h, m_J, Nsample, beta)
        sample_lists = sampleset.record.sample
        
        for i in range(num_spin):
            for j in range(i+1, num_spin):
                sample_ave = np.average([sample_lists[k][i] * sample_lists[k][j] for k in range(len(sample_lists))]) # ã‚µãƒ³ãƒ—ãƒ«ã®çµŒé¨“å¹³å‡
                train_J[(i, j)] += eta * beta * (data_ave[(i, j)] - sample_ave) # ç›¸äº’ä½œç”¨ã®æ›´æ–°

        if (t+1) % 10 == 0 or (t+1) == 1:
            # ã‚µãƒ³ãƒ—ãƒªãƒ³ã‚°
            m_J = minus_J(num_spin=num_spin, J=train_J)
            sampleset = sampling(method=method, h=h, J=m_J, Nsample=150, beta=beta)
            sample_lists = sampleset.record.sample
            # å°¤åº¦è¨ˆç®—
            first = cal_first(beta=beta, J=train_J, data_lists=sample_lists)
            second = cal_second(beta=beta, N=num_spin, J=train_J)
            LH_hists[t+1] = first - second

    return train_J, LH_hists
```

### å³å¯†å‹¾é…ã«ã‚ˆã‚‹å­¦ç¿’

æœ€å¾Œã«ã€æœŸå¾…å€¤ã®è¨ˆç®—ã‚’ã‚µãƒ³ãƒ—ãƒªãƒ³ã‚°ã«ã‚ˆã‚‹è¿‘ä¼¼ã§ã¯ãªãã€å³å¯†ã«è¨ˆç®—ã™ã‚‹ã“ã¨ã§å­¦ç¿’ã‚’è¡Œã„ã¾ã™ã€‚

å‹¾é…æ³•ã‚’ã€ã‚‚ã†ä¸€åº¦è¨˜è¼‰ã—ã¾ã™ã€‚

$$
J_{i j}(t+1)=J_{i j}(t)+\eta \frac{\partial S}{\partial J_{i j}}
$$

$$
\frac{1}{\beta} \frac{\partial S}{\partial J_{i j}}=\left\langle s_{i} s_{j}\right\rangle_{\rho _\mathcal{D}}-\left\langle s_{i} s_{j}\right\rangle_{\rho}
$$

QAã€SAã®å ´åˆã€æœŸå¾…å€¤$\left\langle{}\right\rangle_{\rho}$ã¯

$$
\left\langle s_{i} s_{j}\right\rangle_{\rho} = \frac{1}{L} \sum_{l=1}^{L} s_{i}^{(l)} s_{j}^{(l)}
$$

ã¨ã—ã¦è¿‘ä¼¼çš„ã«æ±‚ã‚ã¾ã—ãŸã€‚ã“ã“ã§ã¯ã€ã“ã®éƒ¨åˆ†ã‚’å³å¯†ã«è¨ˆç®—ã—ã¦ã„ãã¾ã™ã€‚

å…·ä½“çš„ã«ã¯ã€æ¬¡ã®å¼ã§æ±‚ã‚ã‚‰ã‚Œã¾ã™ã€‚

$$
\left\langle s_{i} s_{j}\right\rangle_{\rho}=\sum_{s_{i}=\pm 1} \sum_{s_{j}=\pm 1} s_{i} ãƒ» s_{j} ãƒ» P\left(s_{i}, s_{j}\right)
$$

ã“ã“ã§ã€$P\left(s_{i}, s_{j}\right)$ã¯åŒæ™‚ç¢ºç‡ã§
$$
P\left(s_{i}, s_{j}\right) = \sum_{\mathbf{s} \backslash s_i, s_j} P(\mathbf{s})
$$
ã¨ã—ã¦æ±‚ã‚ã‚‰ã‚Œã¾ã™ã€‚$\mathbf{s} \backslash s_i, s_j$ã€€ã¯ã€$s_i, s_j$ä»¥å¤–å…¨ã¦ã®$\mathbf{s}$ã‚’è¡¨ã—ã¦ã„ã¾ã™ã€‚

ä¾‹)ã€€$N=3, i=1, j=2$ã®å ´åˆ


$$
\left\langle s_{1} s_{2}\right\rangle_{p}=\sum_{s_{1}=\pm 1} \sum_{s_{2}=\pm 1} s_{1} ãƒ» s_{2} ãƒ» P\left(s_{1,} s_{2}\right)
$$

ã“ã“ã§ã€$P\left(s_{1}, s_{2}\right)$ã¯
$$
P\left(s_{1}, s_{2}\right) = \sum_{\mathbf{s} \backslash s_1 s_2} P(\mathbf{s}) = P(s_1, s_2, -1) + P(s_1, s_2, +1)
$$

ã¨ãªã‚Šã¾ã™ã€‚$P(s_1, s_2, s_3)$ã¯ã€ã‚«ãƒãƒ‹ã‚«ãƒ«åˆ†å¸ƒ
$$
P(\mathbf{s})=\frac{1}{Z} \exp (-\beta E(\mathbf{s})), \quad Z=\sum_{\mathbf{s}} \exp (-\beta E(\mathbf{s}))
$$

ã§æ±‚ã‚ã‚‰ã‚Œã¾ã™ã€‚

```python
def cal_statesum(N, J, beta):
    '''åˆ†é…é–¢æ•°ã®è¨ˆç®—'''
    statesum = np.sum([np.exp(- beta * cal_energy(N=N, J=J, spin=spin_list)) for spin_list in itertools.product([-1, 1], repeat=N)])
    return statesum
```

```python
def arg_lists(i, j, N):
    '''åŒæ™‚ç¢ºç‡ã®å¼•æ•°'''
    arg_lists = []
    for z in itertools.product([-1,1], repeat=N-2):
        arg_list = list(z)
        arg_list.insert(i, 'zi')
        arg_list.insert(j, 'zj')
        arg_lists.append(arg_list)
    return arg_lists
```

```python
def canonical(statesum, beta, N, J, arg_list):
    '''ã‚«ãƒãƒ‹ã‚«ãƒ«åˆ†å¸ƒ'''
    energy = cal_energy(N=N, J=J, spin=arg_list)
    canonical = np.exp(- beta * energy) / statesum
    return canonical
```

```python
def joint_prob(N, i, j, arg_lists, zi, zj, ca_dict):
    '''åŒæ™‚ç¢ºç‡'''
    joint_prob=0
    for arg_list in arg_lists:
        arg_list[i], arg_list[j] = zi, zj
        joint_prob += ca_dict[tuple(arg_list)]
    return joint_prob
```

```python
def exact_grad(Tall, beta, N, eta):
    # ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã®åˆæœŸåŒ–
    train_J= {(i, j): 0 for i in range(N) for j in range(i+1, N)}

    # ãƒ‡ãƒ¼ã‚¿ã®çµŒé¨“å¹³å‡
    data_ave={}
    for i in range(N):
        for j in range(i+1, N):
            data_ave[(i, j)] = np.average([data_lists[k][i] * data_lists[k][j] for k in range(len(data_lists))])
    

    LH_hists={}
    for t in tqdm(np.arange(Tall)):
        # å…¨é€šã‚Šã®ã‚«ãƒãƒ‹ã‚«ãƒ«åˆ†å¸ƒã‚’å…ˆã«è¨ˆç®—
        statesum = cal_statesum(N=N, J=train_J, beta=1)
        ca_dict={}
        for spin_list in itertools.product([-1, 1], repeat=N):
            ca_dict[spin_list] = canonical(statesum=statesum, beta=1, N=N, J=train_J, arg_list=spin_list)

        for i in range(N):
            for j in range(i+1, N):
                # å³å¯†ã«æœŸå¾…å€¤ã‚’è¨ˆç®—
                exact_exval=0
                for z in itertools.product([-1,1], repeat=2):
                    zi, zj = z[0], z[1]
                    jp = joint_prob(N=N, i=i, j=j, arg_lists=arg_lists(i, j, N), zi=zi, zj=zj, ca_dict=ca_dict)
                # ç›¸äº’ä½œç”¨ã®æ›´æ–°
                train_J[(i, j)] += eta * beta * (data_ave[(i, j)] - exact_exval)
                
        if (t+1) % 1 == 0 or (t+1) == 1:
            # å°¤åº¦è¨ˆç®—
            first = cal_first(beta=beta, J=train_J, data_lists=data_lists)
            second = cal_second(beta=beta, N=N, J=train_J)
            LH_hists[t+1] = first - second

    return train_J, LH_hists
```

QAã‚„SAã‚ˆã‚Šè¨ˆç®—å›æ•°ãŒå¤šã„ãŸã‚ã€æ™‚é–“ãŒã‹ã‹ã‚Šã¾ã™ã€‚

```python
exact_J, exact_hists = exact_grad(Tall=100, beta=1, N=15, eta=0.01)
```
